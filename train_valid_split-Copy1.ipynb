```python
import os
import glob as glob
import random
import shutil
```

```python
# set the path
image_paths='./dataset/training/image_2/*'
label_paths='./dataset/training/YOLO_label/*'
```

```python
images = glob.glob(image_paths)
labels = glob.glob(label_paths)
images.sort()
labels.sort()
```

```python
iter_num = 1500
min_num = 1
max_num = len(images)
```

```python
img_file_name = 0
img_file_name = random.sample(range(max_num), iter_num)
print(img_file_name)
```

```python
valid_img_path = './dataset/valid/images/'
valid_lbl_path = './dataset/valid/labels/'
for i in img_file_name:
    rand_img_name = './dataset/training/image_2/{:06d}.png'.format(i)
    rand_lbl_name = './dataset/training/lbl_yolo/{:06d}.txt'.format(i)
    dest_img_name = './dataset/valid/images/{:06d}.png'.format(i)
    dest_lbl_name = './dataset/valid/labels/{:06d}.txt'.format(i)
    shutil.move(rand_img_name, dest_img_name)    
    shutil.move(rand_lbl_name, dest_lbl_name)
```
